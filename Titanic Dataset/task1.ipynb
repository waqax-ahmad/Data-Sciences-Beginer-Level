```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('c:/Users/10win/Downloads/titanic.csv')
```

```python
print(df.head())
```

```python
print("Missing Values: \n",df.isnull().sum())
```

```python
df['Age'].fillna(df['Age'].mean(), inplace=True)
df['Embarked'].fillna(df['Embarked'].mode()[0], inplace=True)
print("Missing Values: \n",df.isnull().sum())
```

```python
# df.drop(columns=['Cabin'], inplace=True)
print("Missing Values: \n",df.isnull().sum())
```

```python
# Removin Duplicates

df.drop_duplicates(inplace=True)
print("Duplicates: ", df.duplicated().sum())
```

```python
sns.boxenplot(x=df['Fare'])
plt.title('Boxenplot of Fare')
plt.show()
df = df[df['Fare'] < 300]
plt.show()
```

```python
#Bar chart of Survived Passengers 

sns.countplot(x='Survived', data=df)
plt.title('Survived Count of Passengers')
plt.show()
```

```python
#bar chart of SEX/Gender

sns.countplot(x='Sex', data=df)
plt.title('Gender count of Passengers')
plt.show()
```

```python
#bar chart of Embarked

sns.countplot(x='Embarked', data=df)
plt.title("Embarked count of Passengers")
plt.show()
```

```python
plt.hist(df['Age'], bins=30, edgecolor='black')
plt.title("Age distribution of Passengers")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()
```

```python
plt.hist(df["Fare"], bins=30, edgecolor='black')
plt.title("Fare distribution of Passengers")    
plt.xlabel("Fare")
plt.ylabel("Count")
plt.show()
```

```python
# Selecting only Numeric Columns

numeric_df = df.select_dtypes(include=[np.number])
```

```python
#coorrelation matrix
corr = numeric_df.corr()

#Heatmap of correlation matrix
plt.figure(figsize=(10, 8))
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f')

plt.title("Correlation Heatmap")
plt.show()
```

Summary


In the Titanic dataset, most passengers did not survive, with females showing a significantly higher survival rate than males. The majority of passengers were between 20 and 40 years old. Fare values were highly skewed, with most passengers paying lower fares and a few paying very high fares. The majority of passengers boarded from Southampton. Correlation analysis showed that higher fares and traveling with family members (Parch) were slightly associated with better survival chances, while age and class had weaker or negative correlations with survival.
